```python
import requests
import json
```

```python
#itunesAPI
#documentation can be found at https://developer.apple.com/library/archive/documentation
#specify paramters term, country, media
params = {"term" : "Avril lavigne", "country" : "NL", "media" : "music"}
url = "https://itunes.apple.com/search"
iTunes_response = requests.get(url, params = params)
py_data = json.loads(iTunes_response.text)
```

```python
py_data
```

```python
#a list comprehension that extract the trackname from the list of dictionaries in the higher level dictionary results
songs_avril = [r["trackName"] for r in py_data["results"]]
print(songs_avril)
```

```python
len(songs_avril)
```

```python
album_avril = [r["collectionCensoredName"] for r in py_data["results"]]
```

```python
sng_album = list(zip(songs_avril, album_avril))
```
